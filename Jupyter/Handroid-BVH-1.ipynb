```python
csv_df = pd.read_csv(csv_file)
csv_df.head()
```

```python
df = csv_df[['yaw_x','pitch_z','roll_y']]
df.head()
```

```python
# xl = pd.ExcelFile(bvh_excel)
excel_df = pd.read_excel(bvh_excel, header=[0,1])
# excel_df = excel_df.reset_index(drop=True)
excel_df
```

```python
new_excel_df = pd.DataFrame(columns=excel_df.columns)
new_excel_df.head()
```

Set the roll,pitch,yaw to one of the joints

```python
joint = 'RightShoulder'
new_excel_df[(joint,'Zrotation')] = csv_df['yaw_x']
new_excel_df[(joint,'Yrotation')] = csv_df['roll_y']
new_excel_df[(joint,'Xrotation')] = csv_df['pitch_z']
new_excel_df[[(joint,'Zrotation'),(joint,'Yrotation'),(joint,'Xrotation')]].head(5)
```

```python
hips = [4.3228,16.5922,-15.2362]
right_hip_z_rotation = 20
left_hip_z_rotation = -20
hips
```

```python
new_excel_df[('Hips','Xposition')] = hips[0]
new_excel_df[('Hips','Yposition')] = hips[1]
new_excel_df[('Hips','Zposition')] = hips[2]
new_excel_df[('RightHip','Zrotation')] = right_hip_z_rotation
new_excel_df[('LeftHip','Zrotation')] = left_hip_z_rotation
new_excel_df.fillna(0, inplace=True)
new_excel_df.head()
```

```python
import numpy as np 
np.savetxt(r'c:\temp\np.txt', new_excel_df.values, fmt='%3.4f') #, fmt='%d'
```
